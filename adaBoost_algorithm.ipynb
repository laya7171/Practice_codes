```python
import numpy as np
import pandas as pd
from scipy.stats import zscore
from scipy.stats.mstats import winsorize
```

```python
threshold = 3
numerical_feature = ["Fare", "Age"]
z_scores = np.abs(zscore(df[numerical_feature]))
outliers = np.where(z_scores > threshold)
print("Outliers detected at indices:", outliers)
```

```python
df["Age"] = winsorize(df["Age"], limits=[0.15, 0.15])
df["Fare"] = winsorize(df["Fare"], limits=[0.15, 0.15])
```

```python
print("Outliers detected at indices:", outliers)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder()
```

```python
df["Sex"] = oe.fit_transform(df[["Sex"]])
df["Embarked"] = oe.fit_transform(df[["Embarked"]])
df["Survived"] = oe.fit_transform(df[["Survived"]])
```

```python
df.head()
```

```python
x = df.drop("Survived", axis = 1)
```

```python
y = df["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y , random_state = 42, test_size = 0.2)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ab = AdaBoostClassifier(n_estimators = 45, learning_rate = 1, random_state = 42)
```

```python
ab.fit(X_train, y_train)
```

```python
pred = ab.predict(X_test)
```
