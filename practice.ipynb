```python
def get_parameters(df):
    parameters = {}
    for col in df.columns[df.isnull().any()]:
        if df[col].dtype in ['float64', 'int64', 'int32']:
            strategy = 'mean'
        else:
            strategy = 'most_frequent'
        
        # Handling missing values
        if strategy == 'mean':
            # For numeric columns, use the mean of the column
            missing_value = df[col].mean()
        else:
            # For non-numeric columns, use the most frequent value
            missing_value = df[col].mode()[0]
        
        parameters[col] = {'missing values': missing_value, 'strategy': strategy}
    
    return parameters
get_parameters(df)
```

```python
parameters = get_parameters(df)
```

```python
from sklearn.impute import SimpleImputer

def apply_imputation(df, parameters):
    for col, parm in parameters.items():
        missing_values = parm['missing values']
        strategy = parm['strategy']
        
        if strategy == 'mean':
            imp = SimpleImputer(strategy='mean')
        elif strategy == 'most_frequent':
            imp = SimpleImputer(strategy='most_frequent')
        else:
            # For custom strategies like 'constant'
            imp = SimpleImputer(strategy='constant', fill_value=missing_values)
        
        df[col] = imp.fit_transform(df[[col]])
    
    return df
```
